```python
import pandas as pd
df = pd.read_csv('data.csv')
```

```python
df.price_paid = df.price_paid.apply(lambda x: x.replace("$", ""))
df.price_paid = df.price_paid.astype(float)
```

```python
df.groupby(['gender', 'price_paid']).sum()
```

```python
df.groupby('gender') ['price_paid'].sum()
```

```python
df.groupby(['gender', 'country']).mean()
```

```python
df.groupby('date')['price_paid'].sum()
```

```python
df.groupby('date')['price_paid'].sum().plot(figsize=[20, 10])
```

```python
df.groupby('gender')['price_paid'].sum().plot.pie(legend=True, figsize=[7,7])
```

```python
df.groupby('country')['price_paid'].sum().plot.bar(legend=True, rot=45, figsize=[10,7])
```
