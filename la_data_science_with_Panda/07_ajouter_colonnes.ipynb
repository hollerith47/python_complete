```python
import pandas as pd
df = pd.read_csv("data.csv")
```

```python
df
```

```python
df["tax"].fillna(0, inplace=True)
```

```python
df
```

```python
df.dtypes
```

```python
df.price_paid = df.price_paid.apply(lambda x: x.replace("$", ""))
df.price_paid = df.price_paid.astype(float)
```

```python
df
```

```python
df.dtypes
```

```python
df['price_total'] = df['price_paid'] * (1 - df['tax'] / 100)
```

```python
df
```

```python
countries = {"United States": "UN", "France": "FR", "Canada": "CA", "Marocco": "MA"}
df['country_code']= df['country'].map(countries)
```
